```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
coage = pd.read_csv("COVID-19_Cases_and_Deaths_by_Age_Group_-_ARCHIVE.csv")
```

```python
coage.head()
```

```python
coage.isnull().sum() # get the sum of the null values
```

```python
coage.drop(["Probable cases"], axis=1, inplace=True)
# drop the unwanted column
```

```python
coage.head()
```

```python
values1=["Confirmed cases","Total case rate","Confirmed deaths","Probable deaths"]
coage[values1]=coage[values1].fillna(0)
coage.head()
# fill the null values with zero
```

```python
coage.to_csv("covidagegroup.csv")
```

```python
coage.drop(["Unnamed: 0"], axis=1, inplace=True)
# drop unwanted columns
```

```python
coage
```

```python
coage1=["19-Oct"]
coage=coage[coage.AgeGroups.isin(coage1)==False]
coage
# remove the rows containing the value 19-oct in the column AgeGroups
```

```python
coage.to_csv("covidagegroup.csv")
```

```python
coage.groupby('AgeGroups').sum()['Total cases'].sort_values(ascending = False)
# get the sum of the total cases in each age group
```

```python
sns.barplot(data = coage , x = 'AgeGroups' , y = 'Total cases' , estimator=sum)
#Display total cases in each age group using the barplot graph
```

```python
coage.groupby('AgeGroups').sum()['Total deaths'].sort_values(ascending = False)
# get the sum of the total death in each age group
```

```python
sns.barplot(data = coage , x = 'AgeGroups' , y = 'Total deaths' , estimator=sum)
# Display the total death in each age group using the barplot
```

```python
coage['AgeGroups'] = coage['AgeGroups'].replace('80 and older','80+')
print(coage)
# replace the value 80 and older with 80+ in the Agegroups due to the limited space of the x axis in the plot
```
